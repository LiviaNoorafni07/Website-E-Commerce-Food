```python
# memberi nama kembali pada kolom
df2 = df2.rename(columns = {'Location': 'Lokasi',
                            'Year': 'Tahun',
                            'Distance (in meters)': 'Jarak (Meter/m)',
                            'Stroke': 'Gaya Renang',
                            'Gender': 'Jenis Kelamin',
                            'Team': 'Asal Team',
                            'Athlete': 'Nama Atletik',
                            'Results': 'Waktu',
                            'Rank': 'Peringkat'})
```

```python
df2
```

```python
#mengganti bahasa inggris ke bahasa indonesia pada colom gaya renang
mapping_gaya_renang = {
    'Backstroke': 'Punggung',
    'Breaststroke': 'Dada',
    'Butterfly': 'Kupu-kupu',
    'Freestyle': 'Gaya Bebas',
    'Individual medley': 'Individu Medley'
}

# Mengganti bahasa Inggris ke dalam bahasa Indonesia pada kolom 'Gaya Renang'
df2['Gaya Renang'] = df2['Gaya Renang'].map(mapping_gaya_renang)
```

```python
df2['Jenis Kelamin'] = df2['Jenis Kelamin'].replace({'Men': 'Laki-laki', 'Women': 'Perempuan'})
```

```python
df2
```

```python
#menambah ID
df2['ID Atletik'] = df2.reset_index(drop=True).index + 1
```

```python
#mengurutkan colom
df3 = df2[['ID Atletik', 'Waktu','Nama Atletik', 'Jenis Kelamin', 'Asal Team', 'Gaya Renang', 'Jarak (Meter/m)', 'Peringkat', 'Tahun', 'Lokasi']]
df3
```

```python
def rapikan_waktu(waktu):
    if pd.notna(waktu):
        parts = waktu.split(':')
        if len(parts) == 2:
            # Tambahkan "00:" pada bagian jam dan rapikan detik
            return f'00:{parts[0]}:{parts[1]}'
        elif len(parts) == 4:
            # Rapikan format waktu yang memiliki jam, menit, detik, dan mikrodetik
            return f'{parts[0]}:{parts[1]}:{parts[2]}'
        else:
            return 'Format waktu tidak valid'
    else:
        return waktu

# Menggunakan fungsi untuk merapikan waktu pada setiap baris data
df3['Waktu'] = df3['Waktu'].apply(rapikan_waktu)

# Mengubah kolom 'Waktu' ke dalam format yang sesuai
df3['Waktu'] = pd.to_datetime(df3['Waktu'], format='%H:%M:%S', errors='coerce').dt.time
```

```python
df3 = df3.dropna(subset=['Waktu'])
df3
```
